```python
import pandas as pd
```

```python
psl_df= pd.read_csv('psl.csv')
```

```python
psl_df
```

```python
psl_total_score= psl_df.groupby(['match_id', 'inning']).sum()['total_runs'].reset_index()
```

```python
psl_total_score= psl_total_score[psl_total_score['inning']== 1]
psl_total_score['total_runs']= psl_total_score['total_runs'] + 1
```

```python
psl_total_score.head()
```

```python
psl_df= psl_total_score.merge(psl_df, on= 'match_id')
```

```python
psl_df= psl_df[psl_df['inning_y']== 2]
```

```python
psl_df.columns
```

```python
psl_df['current_score']= psl_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
psl_df['runs_left']= psl_df['total_runs_x'] - psl_df['current_score']
```

```python
psl_df['balls_left']= 126 - (psl_df['over']*6  + psl_df['ball'])
```

```python
psl_df['is_wicket']= psl_df['is_wicket'].replace(True, 1)
psl_df['is_wicket']= psl_df['is_wicket'].replace(False, 0)
```

```python
wickets = psl_df.groupby('match_id')['is_wicket'].cumsum()
psl_df['wickets_left'] = 10 - wickets
psl_df.head()
```

```python
# CRR & RRR
psl_df['crr']= (psl_df['current_score']*6)/ (120- psl_df['balls_left'])
psl_df['rrr']= (psl_df['runs_left']*6)/ psl_df['balls_left']
```

```python
psl_df['batting_team'].unique()
```

```python
def res(row):
    return 1 if row['batting_team']== row['winner'] else 0
```

```python
psl_df['result']= psl_df.apply(res, axis= 1)
```

```python
psl_df.columns
```

```python
df= psl_df[['batting_team', 'bowling_team', 'venue', 'runs_left', 'balls_left', 'wickets_left', 'total_runs_x', 'crr', 'rrr', 'result']]
```

```python
df= df.sample(df.shape[0])
```

```python
df.isna().sum()
```

```python
df.dropna(inplace= True)
```

```python
df= df[df['balls_left'] != 0]
```

```python
df
```

```python
X= df.iloc[:,:-1]
y= df.iloc[:,-1]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.30, random_state=0)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
trf= ColumnTransformer([
    ('ohe', OneHotEncoder(sparse_output= False, drop='first'), ['batting_team', 'bowling_team', 'venue']),
    ('sc', StandardScaler(), ['runs_left', 'balls_left', 'wickets_left', 'total_runs_x', 'crr', 'rrr'])
], remainder= 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe= Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver= 'liblinear'))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred= pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
