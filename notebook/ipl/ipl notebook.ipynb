# removing old teams

```python
match_df['team1'].unique()
```

```python
teams= [
    'Mumbai Indians', 
    'Kolkata Knight Riders',
    'Rajasthan Royals',
    'Chennai Super Kings',
    'Sunrisers Hyderabad',
    'Delhi Capitals',
    'Punjab Kings',
    'Lucknow Super Giants', 
    'Gujarat Titans',
    'Royal Challengers Bengaluru'
]
```

```python
match_df['team1']= match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2']= match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['winner']= match_df['winner'].str.replace('Delhi Daredevils', 'Delhi Capitals')

match_df['team1']= match_df['team1'].str.replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru')
match_df['team2']= match_df['team2'].str.replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru')
match_df['winner']= match_df['winner'].str.replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru')

match_df['team1']= match_df['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team2']= match_df['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['winner']= match_df['winner'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')

match_df['team1']= match_df['team1'].str.replace('Kings XI Punjab', 'Punjab Kings')
match_df['team2']= match_df['team2'].str.replace('Kings XI Punjab', 'Punjab Kings')
match_df['winner']= match_df['winner'].str.replace('Kings XI Punjab', 'Punjab Kings')
```

```python
match_df= match_df[match_df['team1'].isin(teams)]
match_df= match_df[match_df['team2'].isin(teams)]
```

```python
match_df.head()
```

```python
match_df.shape
```

# taking only matches with no rain

```python
match_df['method']= match_df['method'].fillna(0)
```

```python
match_df= match_df[match_df['method']== 0]
```

```python
match_df= match_df[['match_id', 'city', 'winner', 'total_runs']]
```

```python
match_df.head()
```

```python
delivery_df= match_df.merge(deli, on= 'match_id')
```

```python
delivery_df= delivery_df[delivery_df['inning']== 2]
```

```python
delivery_df['current_score']= delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df['runs_left']= delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df['balls_left']= 120 - (delivery_df['over']*6  + delivery_df['ball'])
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
```

```python
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum()
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
# CRR & RRR
delivery_df['crr']= (delivery_df['current_score']*6)/ (120- delivery_df['balls_left'])
delivery_df['rrr']= (delivery_df['runs_left']*6)/ delivery_df['balls_left']
```

```python
delivery_df['batting_team']= delivery_df['batting_team'].str.replace('Delhi Daredevils', 'Delhi Capitals')
delivery_df['bowling_team']= delivery_df['bowling_team'].str.replace('Delhi Daredevils', 'Delhi Capitals')

delivery_df['batting_team']= delivery_df['batting_team'].str.replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru')
delivery_df['bowling_team']= delivery_df['bowling_team'].str.replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru')

delivery_df['batting_team']= delivery_df['batting_team'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
delivery_df['bowling_team']= delivery_df['bowling_team'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')

delivery_df['batting_team']= delivery_df['batting_team'].str.replace('Kings XI Punjab', 'Punjab Kings')
delivery_df['bowling_team']= delivery_df['bowling_team'].str.replace('Kings XI Punjab', 'Punjab Kings')

delivery_df['city']= delivery_df['city'].str.replace('Bangalore', 'Bengaluru')
```

```python
delivery_df= delivery_df[delivery_df['batting_team'].isin(teams)]
delivery_df= delivery_df[delivery_df['bowling_team'].isin(teams)]
```

```python
delivery_df.sample()
```

```python
def res(row):
    return 1 if row['batting_team']== row['winner'] else 0
```

```python
delivery_df['result']= delivery_df.apply(res, axis= 1)
```

```python
df= delivery_df[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left', 'wickets', 'total_runs_x', 'crr', 'rrr', 'result']]
```

```python
df= df.sample(df.shape[0])
```

```python
df.dropna(inplace= True)
```

```python
df= df[df['balls_left'] != 0]
```

```python
df['result'].value_counts()
```

```python
df
```

# Model Training

```python
X= df.iloc[:,:-1]
y= df.iloc[:,-1]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.30, random_state=0)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
trf= ColumnTransformer([
    ('ohe', OneHotEncoder(sparse_output= False, drop='first'), ['batting_team', 'bowling_team', 'city']),
    #('sc', StandardScaler(), ['runs_left', 'balls_left', 'wickets', 'total_runs_x', 'crr', 'rrr'])
], remainder= 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe= Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver= 'liblinear'))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred= pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
toss_df= toss_df[toss_df['team1'].isin(teams)]
toss_df= toss_df[toss_df['team2'].isin(teams)]
```

```python
toss_df= toss_df[~toss_df['match_type'].isin(['Elimination Final', '3rd Place Play-Off', 'Semi Final'])]
```

```python
Xt= toss_df.iloc[:,:-1]
yt= toss_df.iloc[:,-1]

from sklearn.model_selection import train_test_split
Xt_train, Xt_test, yt_train, yt_test= train_test_split(Xt, yt, test_size= 0.20, random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
trft= ColumnTransformer([
    ('ohe', OneHotEncoder(sparse_output= False, drop='first'), ['team1', 'team2', 'city', 'match_type'])
], remainder= 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe_t= Pipeline(steps=[
    ('step1', trft),
    ('step2', RandomForestClassifier(random_state=42))
])
```

```python
pipe_t.fit(Xt_train, yt_train)
```

```python
yt_pred= pipe_t.predict(Xt_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(yt_test, yt_pred)
```
