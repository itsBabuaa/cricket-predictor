```python
target_score_df= odi_ball.groupby(['id', 'innings'])['total_runs'].sum().reset_index()
```

```python
target_score_df= target_score_df[target_score_df['innings']== 1]
```

```python
target_score_df['total_runs']= target_score_df['total_runs'] + 1
```

```python
target_score_df
```

```python
match_df= odi_match.merge(target_score_df[['id', 'total_runs']], left_on= 'id', right_on= 'id')
```

```python
match_df.head()
```

```python
match_df['team1'].unique()
```

```python
# active teams
teams= ['New Zealand', 'India', 'Sri Lanka', 'Australia', 'England',
       'West Indies', 'Zimbabwe', 'Canada', 'Netherlands',
       'Pakistan', 'South Africa', 'Bangladesh','United States of America',
        'Ireland', 'Scotland', 'Hong Kong', 'Afghanistan',
        'United Arab Emirates', 'Namibia', 'Oman', 'Nepal', 'Thailand']
```

```python
match_df= match_df[match_df['team1'].isin(teams)]
match_df= match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df['method'].unique()
```

```python
#only matches with no rain
match_df['method']= match_df['method'].fillna(0)
match_df= match_df[match_df['method']== 0]
```

```python
match_df= match_df[['id','gender', 'city', 'winner', 'total_runs']]
```

```python
match_df
```

```python
odi_df= match_df.merge(odi_ball, on= 'id')
```

```python
odi_df= odi_df[odi_df['innings']== 2]
```

```python
odi_df.info()
```

```python
odi_df['current_score']= odi_df.groupby('id')['total_runs_y'].cumsum()
```

```python
odi_df['runs_left']= odi_df['total_runs_x'] - odi_df['current_score']
```

```python
odi_df= odi_df[~odi_df['ball'].isin([7,8,9,10,11,12])]
```

```python
odi_df['balls_left']= 300 - (odi_df['over']*6  + odi_df['ball'])
```

```python
odi_df['is_wkt_delivery'].unique()
```

```python
wickets = odi_df.groupby('id')['is_wkt_delivery'].cumsum()
odi_df['wickets_left'] = 10 - wickets
odi_df.head()
```

```python
# CRR & RRR
odi_df['crr']= (odi_df['current_score']*6)/ (300- odi_df['balls_left'])
odi_df['rrr']= (odi_df['runs_left']*6)/ odi_df['balls_left']
```

```python
odi_df.shape
```

```python
def res(row):
    return 1 if row['batting_team']== row['winner'] else 0
```

```python
odi_df['result']= odi_df.apply(res, axis= 1)
```

```python
df= odi_df[['gender', 'batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left', 'wickets_left', 'total_runs_x', 'crr', 'rrr', 'result']]
```

```python
df.sample()
```

```python
df= df.sample(df.shape[0])
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df= df[df['balls_left'] != 0]
```

```python
df['result'].value_counts()
```

# ODI MEN

```python
men_df= df[df['gender']== 'male']
```

```python
men_df.head()
```

```python
men_df= men_df[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left', 'wickets_left', 'total_runs_x', 'crr', 'rrr', 'result']]
```

```python
men_df
```

```python
X= men_df.iloc[:,:-1]
y= men_df.iloc[:,-1]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
trf= ColumnTransformer([
    ('ohe', OneHotEncoder(sparse_output= False, drop='first'), ['batting_team', 'bowling_team', 'city']),
    ('sc', StandardScaler(), ['runs_left', 'balls_left', 'wickets_left', 'total_runs_x', 'crr', 'rrr'])
], remainder= 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
men_pipe= Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver= 'liblinear'))
])
```

```python
men_pipe.fit(X_train, y_train)
```

```python
y_pred= men_pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, r2_score
print(accuracy_score(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
